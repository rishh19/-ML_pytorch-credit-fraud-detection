# Credit Card Fraud Detection using PyTorch

## Project Overview
Credit card fraud is a major issue in financial systems. Fraudulent transactions cause significant financial losses and undermine trust in digital payment systems. Machine learning can help detect suspicious transactions by identifying unusual patterns in transaction data.

In this project, we build a **deep learning model using PyTorch** to classify whether a transaction is fraudulent or legitimate.

---

## Problem Statement
Financial institutions process millions of transactions daily. Among these, a very small percentage are fraudulent. Detecting fraud is challenging because:

- Fraudulent transactions are extremely rare
- Fraud patterns constantly change
- False positives must be minimized to avoid blocking legitimate transactions

The goal of this project is to build a **machine learning model capable of identifying fraudulent credit card transactions with high accuracy and recall**.

---

## Objectives
The main objectives of this project are:

- Perform exploratory data analysis (EDA) on the dataset
- Handle the class imbalance problem
- Preprocess and scale transaction features
- Build a neural network using PyTorch
- Train and evaluate the model
- Deploy the model for real-time fraud prediction

---

## Dataset
The dataset used in this project is the **Credit Card Fraud Detection Dataset** available on Kaggle.

Dataset characteristics:

- Transactions made by European cardholders
- Highly imbalanced dataset
- Fraud cases represent only a small fraction of total transactions

Features:
- V1 to V28 : anonymized transaction features
- Time : time elapsed between transactions
- Amount : transaction amount
- Class : target variable (0 = normal, 1 = fraud)
